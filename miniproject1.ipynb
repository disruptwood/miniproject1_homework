```python
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
def calculateExtreme(arr: np.array):
    '''
    Find the indices of local maxima and minima in the array.
    '''
    slope = np.diff(arr)  # Calculate the slope between points

    # Find where slope changes from positive to negative (maxima)
    maxima = np.where((slope[:-1] > 0) & (slope[1:] < 0))[0] + 1

    # Find where slope changes from negative to positive (minima)
    minima = np.where((slope[:-1] < 0) & (slope[1:] > 0))[0] + 1

    return maxima, minima  # Return indices of maxima and minima

def sinusoidalSignal(arr: np.array, factor=5):
    '''
    Downsample the sinusoidal signal while preserving all maxima and minima.
    '''
    steps = factor
    size = max(len(arr) // steps, 1)

    maxima, minima = calculateExtreme(arr)

    sampledPoints = np.linspace(0, len(arr) - 1, size, dtype=int)
    all_indices = np.unique(np.concatenate((sampledPoints, maxima, minima)))

    sampledArr = arr[all_indices]

    return sampledArr, all_indices
```

```python
# Generate a time array from 0 to 1 second with a step size of 0.01 seconds (100 points)
time = np.arange(0, 1, 0.01)

# Set the frequency of the sinusoidal signal to 5 Hz
frequency = 5  

# Create the sinusoidal signal using the sine function
arr = np.sin(2 * np.pi * frequency * time)

# Downsample the signal while preserving all maxima and minima
sampleArr, sampleIndices = sinusoidalSignal(arr, factor=5)

# Map the sampled indices to their corresponding time values
newTime = time[sampleIndices]
```

```python
plt.figure(figsize=(20, 10))

# Plot the original sinusoidal signal
plt.plot(time, arr, label="Original signal", color='blue', linewidth=2, alpha=0.7)

# Plot the downsampled signal
plt.plot(newTime, sampleArr, 'o-', label="Downsampled signal", color='red', linewidth=2, markersize=8)

# Highlight the maxima and minima on the downsampled signal
differences_down = np.diff(sampleArr)
ds_maxima = np.where((differences_down[:-1] > 0) & (differences_down[1:] < 0))[0] + 1
ds_minima = np.where((differences_down[:-1] < 0) & (differences_down[1:] > 0))[0] + 1

# Scatter plot for maxima
plt.scatter(newTime[ds_maxima], sampleArr[ds_maxima], color='green', s=450, marker='^', label='Maxima')

# Scatter plot for minima
plt.scatter(newTime[ds_minima], sampleArr[ds_minima], color='magenta', s=450, marker='v', label='Minima')

# Add title and labels with larger font sizes for clarity
plt.title("5Hz sinusoid: Original vs Downsampled", fontsize=24)
plt.xlabel("Time (s)", fontsize=20)
plt.ylabel("Amplitude", fontsize=20)

# Add grid for better readability
plt.grid(True)

# Add legend with larger font size
plt.legend(fontsize=16)

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
```

```python
# Load the dataset
# It is in the same directory as the Jupyter notebook
df = pd.read_csv('laptop_price - dataset.csv')

# Display the first few rows
df.head()
```

```python
plt.figure(figsize=(20, 8))
# Create a scatter plot where each laptop is a point
# The color of each point is determined by its price, mapping lower prices to green and higher to red
scatter = plt.scatter(
    x=df.index, 
    y=df['Price (Euro)'], 
    c=df['Price (Euro)'], 
    cmap='RdYlGn_r',  # Reversed Red-Yellow-Green colormap
    s=300,
    edgecolor='k'
)

# Set plot titles and labels
plt.title('Price of all laptops', fontsize=26)
plt.xlabel('Laptop index', fontsize=22)
plt.ylabel('Price (Euro)', fontsize=22)
plt.tick_params(axis='x', labelsize=18)
plt.tick_params(axis='y', labelsize=18)

# Add grid
plt.grid(True)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Display
plt.show()
```

```python
# Group the data by 'Company' and calculate the mean price
average_price = df.groupby('Company')['Price (Euro)'].mean().sort_values(ascending=False)

# Display the average prices
print("Average Laptop Price by Company (in Euro):")
print(average_price)

# Identify the company with the highest average price
most_expensive_company = average_price.idxmax()
highest_average_price = average_price.max()

print(f"\nCompany with the most expensive laptops on average: {most_expensive_company}")
print(f"Average Laptop Price for {most_expensive_company}: ${highest_average_price:.2f}")
```

```python
# Display unique operating systems before standardization
print("Unique Operating Systems before standardization:")
print(df['OpSys'].unique())

# Define a mapping dictionary to standardize OS names
os_mapping = {
    'Windows 10 S': 'Windows',
    'Windows 10': 'Windows',
    'Windows 7': 'Windows',
    'macOS': 'MacOS',
    'Mac OS X': 'MacOS',
    'mac OS X': 'MacOS',
}

# Apply the mapping to the 'OpSys' column
df['OpSys'] = df['OpSys'].map(os_mapping).fillna(df['OpSys'])

# Display unique operating systems after standardization
print("\nUnique Operating Systems after standardization:")
print(df['OpSys'].unique())
```

```python
# Get the unique operating systems
unique_os = df['OpSys'].unique()

# Determine the number of unique OS types for subplot arrangement
num_os = len(unique_os)

# Set up the matplotlib figure with appropriate size
plt.figure(figsize=(20, 5 * num_os))

# Iterate through each operating system and create a subplot
for i, os in enumerate(unique_os, 1):
    plt.subplot(num_os, 1, i)
    sns.histplot(df[df['OpSys'] == os]['Price (Euro)'], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of laptop prices for {os}', fontsize=18)
    plt.xlabel('Price (Euro)', fontsize=21)
    plt.ylabel('Num of laptops', fontsize=21)
    plt.tick_params(axis='x', labelsize=18)
    plt.tick_params(axis='y', labelsize=18)

# Adjust layout
plt.tight_layout()
plt.show()
```
